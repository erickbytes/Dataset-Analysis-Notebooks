```python
# https://data.world/awram/food-nutritional-values
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel('ABBREV.xlsx')
df = df.sort_values(by=['Sugar_Tot_(g)'], ascending=False)
df = df[df.Shrt_Desc.str.contains('COFFEE,')]
df = df.nlargest(10, 'Sugar_Tot_(g)')
df.plot.barh(x='Shrt_Desc', y='Sugar_Tot_(g)', color='DarkBlue', legend=False)
plt.style.use('seaborn-pastel')
plt.xlabel(xlabel='Grams of Sugar')
plt.title('Most Sugar, Coffee - USDA', fontsize=14)
plt.gca().invert_yaxis()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel('ABBREV.xlsx')
df = df.sort_values(by=['Sugar_Tot_(g)'], ascending=True)
df = df[(df.Shrt_Desc.str.contains('WINE') | df.Shrt_Desc.str.contains('BEER')) & (df.Shrt_Desc.str.contains('BEV')) & (~df.Shrt_Desc.str.contains('NON-ALCOHOLIC'))] #
df = df.nlargest(12, 'Sugar_Tot_(g)')
df.plot.barh(x='Shrt_Desc', y='Sugar_Tot_(g)',color='DarkBlue', legend=False)
plt.gca().invert_yaxis()
plt.title('Most Sugar, Beer or Wine Beverages - USDA', fontsize=14)
plt.savefig('Most Sugar, Beer or Wine Beverages - USDA.png', bbox_inches='tight')
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel('ABBREV.xlsx')
df = df.sort_values(by=['Iron_(mg)'], ascending=True)
df = df[(~df.Shrt_Desc.str.contains('BABYFOOD'))] # Filter out baby foods
df = df.nlargest(50, 'Iron_(mg)')
df.plot.barh(x='Shrt_Desc', y='Iron_(mg)',color='DarkBlue', legend=False, figsize=(15,10))
plt.gca().invert_yaxis()
plt.style.use('seaborn-pastel')
plt.xlabel(xlabel='Milligrams of Iron')
ax1 = plt.axes()
ax1.yaxis.label.set_visible(False)
plt.title('Foods With The Most Iron - USDA (2015)', fontsize=14)
plt.savefig('Foods with The Most Iron - USDA 2015.png', bbox_inches='tight')
```
